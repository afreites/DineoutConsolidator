```python
tor_path= "Coordinates_datasets/torontocoordinates.csv" 
tor_data=pd.read_csv('Coordinates_datasets/torontocoordinates.csv', index_col=0)
tor_data = pd.read_csv(tor_path)

sample_1 = tor_data.sample(5000)
sample_tor = sample_1.sample(1000)
```

```python
name= []
place_id= []
rating_place = []
user_ratings_total=[]
lat_place =[]
lng_place = []
price_level= []

# ethnicity=[]
for index, row in sample_tor.iterrows():
    # get lat, lng from df
    lat = row["lat"]
    lng = row["lng"]
    gkey= gkey
    coordinates=f"{lat},{lng}"
    target_url =(f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={coordinates}&radius=1500&type=restaurant&keyword=dine&key={gkey}")
    response = requests.get(target_url)
    response_json=response.json()
    results = response_json['results']
    for item in results:
        try:
            name.append(item["name"])
            place_id.append(item["place_id"])
            lat_place.append(item['geometry']['location']['lat'])
            lng_place.append(item['geometry']['location']['lng'])
            user_ratings_total.append(item ["user_ratings_total"])
            rating_place.append(item["rating"])
            price_level.append(item["price_level"])
        except:
            price_level.append(0) 
```

```python
tor_df = pd.DataFrame({"name":name,
                       "place_id":place_id,
                       "rating_place":rating_place,
                       "price_level":price_level,
                       "user_ratings_total": user_ratings_total,
                       "lat_place": lat_place,
                       "lng_place":lng_place})
tor_df.reset_index(drop=True,inplace=True)
```

```python
tor_df.to_csv('toronto_restaurants_google.csv',index = False, header = True)
```

```python
name_det= []
place_id_det= []
formatted_address = []
formatted_phone_number=[]
lat_place_det =[]
lng_place_det = []
reviews = []
url = []
website = []

for index, row in tor_df.iterrows():
    # get lat, lng from df
    place = row["place_id"]
    gkey= gkey
    target_url_details=(f"https://maps.googleapis.com/maps/api/place/details/json?placeid={place}&key={gkey}")
    response_details = requests.get(target_url_details)
    response_json_details=response_details.json() 
    results = response_json_details['result']
    
    for item in results:
        try:
            name_det.append(item["name"])
            place_id_det.append(item["place_id"])
            formatted_address.append(item["formatted_address"])
            formatted_phone_number.append(item["formatted_phone_number"])
            lat_place_det.append(item['geometry']['location']['lat'])
            lng_place_det.append(item['geometry']['location']['lng'])
            reviews.append(item ["reviews"])
            url.append(item["url"])
            website.append(item["website"])
        except:
            formatted_phone_number.append(0) 
```
