```python
# Dependencies and Setup
import requests
import json
import pandas as pd
import time
```

```python
#Declare variables
id_no = []
name = []
price_range = []
review = []
phone = []
reservations = []
rating = []
start=41
count=60
counter = 20
```

```python
api = 'e61d98604aa556372466d466c4f086df'
headers = {'user-key': api}
```

```python
for c in range(2):
    url = f'https://developers.zomato.com/api/v2.1/search?entity_id=89&entity_type=city&start={start}&count={count}'
    print(url)
    response = requests.get(url, headers=headers)
    toronto= json.loads(response.text)

    #add a 5 second interval between queries
    time.sleep(5)
    for r in toronto['restaurants']:
        #print(restaurant)
        restaurant=r['restaurant']
        id_no.append(restaurant['id'])
        name.append(restaurant['name'])
        price_range.append(restaurant['price_range'])
        phone.append(str(restaurant['phone_numbers']))
        reservations.append(str(restaurant['has_table_booking']))
        review.append(restaurant['user_rating']['votes'])
        rating.append(restaurant['user_rating']['aggregate_rating'])
    start = start+20
    count = count+20
```

```python
toronto_df= pd.DataFrame(data={"name":name, "id": id_no, "phone": phone, "reservations": reservations, 
                              "price_range": price_range, "review": review, "rating": rating })
toronto_df=toronto_df.sort_values('id')
toronto_df.head(25)
```
