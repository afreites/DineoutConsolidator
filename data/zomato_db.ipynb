```python
# Dependencies and Setup
import requests
import json
import pandas as pd
import time
from zkeys import zkey
```

```python
#Declare variables
id_no = []
name = []
price_range = []
review = []
phone = []
reservations = []
rating = []
counter = 20
```

```python
#Assign API key
api = zkey
headers = {'user-key': api}
#Toronto neighbourhoods 
entity_ids = (87071, 87081, 87091, 87161, 87191, 87221, 87241, 87281, 87351, 
              87361, 87391, 87401, 87441, 87701, 87861, 87871, 87881, 89161, 87891)
```

```python
#Loop through subzones
for entity_id in entity_ids:
    start=1
    count=20
#Loop through two pages of results
    for c in range(4):
        url = f'https://developers.zomato.com/api/v2.1/search?entity_id={entity_id}&entity_type=subzone&start={start}&count={count}'
        response = requests.get(url, headers=headers)
        response_df= json.loads(response.text)
        #print(url)
        #add a one second interval between queries
        time.sleep(1)

        #Populate the dataframe
        for r in response_df['restaurants']:
            restaurant=r['restaurant']
            id_no.append(restaurant['id'])
            name.append(restaurant['name'])
            price_range.append(restaurant['price_range'])
            phone.append(str(restaurant['phone_numbers']))
            reservations.append(str(restaurant['has_table_booking']))
            review.append(restaurant['user_rating']['votes'])
            rating.append(restaurant['user_rating']['aggregate_rating'])

        #set the next page to pull
        start = start+20
        count = count+20
```

```python
zomato_df= pd.DataFrame(data={"name":name, "id": id_no, "phone": phone, "reservations": reservations, 
                              "price_range": price_range, "review": review, "rating": rating })
zoamto_df=zomato_df.sort_values('id')
zomato_df.head(25)
len(zomato_df)
```
