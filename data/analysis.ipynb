```python
# Dependencies, Setup, API Key
import requests
import json
import pandas as pd
import time
from pprint import pprint
from ykeys import ykey
```

```python
consolidated_LA = "consolidated_LA.csv"
consolidated_df_LA=pd.read_csv(consolidated_LA, encoding="utf-8-sig")
consolidated_df_LA.head(10)
```

```python
#Calculate a composite score for LA
composite_score = []
for score in consolidated_df_LA:
    z_score= consolidated_df_LA['rating']
    y_score= consolidated_df_LA['y_rating']
    g_score = consolidated_df_LA['g_rating']
    composite_score = (z_score+y_score+g_score)/3
```

```python
consolidated_df_LA['composite_score'] = composite_score
consolidated_df_LA.head()
```

```python
#Save to a CSV all data retrieved
consolidated_df_LA.to_csv("consolidated_LA.csv", sep=',',index=False)
```
